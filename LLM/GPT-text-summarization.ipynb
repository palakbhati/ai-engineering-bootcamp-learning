### Summarising Text

```python
def text_summarizer(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {
                "role": "system",
                "content": "you will be provided with thw a blank of text"
            },
            {
                "role": "user",
                "content": "a flying saucer seen by a guest house a seven fit alien like figure coming out of a hedge"
                
            },
            {
                "role": "assistent",
                "content": "a flying saucer seen by a guest house a seven fit alien like figure coming out of a hedge"
                
            },
            {
                "role": "user",
                "content": "a flying saucer seen by a guest house a seven fit alien like figure coming out of a hedge"
                
            },
            {
                "role": "assistent",
                "content": "a flying saucer seen by a guest house a seven fit alien like figure coming out of a hedge"
                
            },
            {
                "role":"user",
                "content":prompt
            }
        ],
        temperature=0.5,
        max_tokens=756
    )
    return response.choices[0].messages.content.strip()
```

```python
prompt="Master reef guide kristy whitem didnt need to tell me twice, peering down through my smorkel mask in the direction"
print(prompt)
```

```python
text_summarizer(prompt)
```

### Poetic chatbot

```python
def poetic_chatbot(prompt):
    response = openai.ChatCompletion.create(
        model = "gpt-3.5-turbo",
        messages = [
            {
                "role": "system",
                "content": "you are a poetic chatbot"
            },
            {
                "role": "user",
                "content": "when was google founded"
            },
            {
                "role": "assistent",
                "content": "i am palak bhati i am curretly learning from a ai engineer bootcamp i am a student"
            },
            {
                "role": "user",
                "content" : "which country has the youngest president"
            },
            {
                "role": "assistent",
                "content" : "which country has the youngest president"
            },
            {
                "role": "user",
                "content":prompt
            }
        ],
        temperature = 1,
        max_tokens=256
    )
    return response.choices[0].message.content.strip()
```

```python
prompt = "when was chiz first made"
poetic_chatbot(prompt)
```
