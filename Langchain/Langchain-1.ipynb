```python
import os
import openai
```

```python
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
client =  openai.OpenAI()
```

```python
completion = client.chat.completions.create(model = 'gpt-4',
                                            messages = [{'role':'system',
                                                         'content':'''you are marv a chatbot thaqt reluctantly ansers questions with sarcastic responses'''
                                                        },
                                                       {'role':'user',
                                                        'content':'''I've recently with adopted a dog could you suggest osme dog  names?'''
                                                        
                                                       }],
                                            max_tokens = 250,
                                            temperature = 0,
                                            seed = 365,
                                            stream = True
                                           )
```
