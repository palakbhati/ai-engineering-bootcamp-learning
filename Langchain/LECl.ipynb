## Piping a prompt , model , an output parser

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import CommaSeparatedListOutputParser
```

```python
list_instructions = CommaSepratedListOutputParser().get_format_instructons()
```

```python
list_instructions
```

```python
chat_template = ChatPromptTemplate.from_messages([('human', "i've recently adopted a {pet}. could you suggest three {pet} names? \n" + list_insturctions)])
```

```python
print(chat_template.messages[0].prompt.template)
```

```python
chat = ChatOpenAI(model_name = 'gpt-4',
                  model_kwargs = {'seed': 365},
                  temperature = 0,
                  max_tokens = 100)
```

```python
list_output_parser = CommaSepratedListOutputParser()
```

```python
chat_template_result = chat_template.invoke({'pet':'dog'})
```

```python
chat_result = chat.invoke(chat_template_result)
```

```python
list_output_parser.invoke(chat_result)
```

```python
chain = chat_template | chat | list_output_parser
```

```python
chain.invoke({'pet' : 'dog'})
```

```python
chat_template = ChatPromptTemplate.from_messages(['human', "I've recently adopted a {pet} which is a {breed}. could you sugget several training tips?")])
```

```python
chat = ChatOpenAI(model_name = 'gpt-4',
                  model_kwargs = {'seed': 365},
                  temperature = 0,
                  max_tokens = 100)
```

```python
chain = chat_template | chat
```

```python
chain.invoke({'pet':'dog' , 'breed':'shephered'})
```

```python
%%time
chain.batch([{'pet':'dog' , 'breed':'shepherd'},{'pet':'dragon' , 'breed':'night fury'}])
```

```python
%%time
chain.invoke({'pet':'dog' , 'breed':'shephered'})
```

# RunnableLambda

```python
find_sum = lambda x: sum(x)
```

```python
find_sum([1,2,5])
```

```python
find_squares = lambda x: x**2
```

```python
find_squares(8)
```

```python
runnable_sum = RunnableLambda(lambda x: sum(x))
```

```python
runnable_sum.invoke([1,2,3])
```

```python
runnable_square = RunnableLambda(lambda x: x**2)
```

```python
runnable_square.invoke()
```

```python
chain = runnable_sum | runnable_square
```
