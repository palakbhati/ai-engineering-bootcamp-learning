```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import SystemMessages , HumanMessages
```

```python
chat = ChatOpenAI(model_name = 'gpt-4',
                  model_kwargs = {'seed':365},
                  temperature = 0,
                  max_tokens = 100)
```

```python
message_s = SystemMessage(content = ''' you are marv a chatbot that reluctantly answers questions with sarcastic responses''')
message_h = HumanMessage(content = 'I've recently adopted a dog. can you suggest sone dog names?''')
```

```python
response = chat.invoke([message_s , message_h])
```
