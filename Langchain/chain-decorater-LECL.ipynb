```python
from langchain_core.runnables import RUnnableLambda
```

```python
def find_sum(x):
    return sum(x)

def find_square(x):
    return x**2
```

```python
chain1 = RunnableLambda(find_sum) | RunnableLambda(find_square)
```

```python
chain1.invoke()
```

```python
@chain
def runnable_sum(x):
    return sum(x)

def runnable_square(x):
    return x**2
```
