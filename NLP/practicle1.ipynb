```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer , WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data = pd.read_csv('tripadvisor_hotel_reviews.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

### Lowercase

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

## stop words removal

```python
en_stopwords = stopwords.words("english")
en_stopwords.remove("not")
```

```python
data['review_no_stopwords'] = data['review_lowercase'] \
                                .apply(lambda x: ' ' \
                                .join([word for word \
                                in x.split() if word \
                                not in (en_stopwords)]))


```

```python
data.head()
```

```python
data['review_no_stopwords'][0]
```

#### punctuation

 We want to remove punctation but we notice we have a few reviews with the symbol * instead of the word star. We want to keep this as it might add meaning to the review. We can do this using what we learned about regular expres


```python
data['review_no_stopwords_no_punct'] = data \
                                    .apply(lambda x: \
                                    re.sub(r"[*]", \
                                    "star", \
                                    x['review_no_stopwords'] \
                                    ), axis=1)
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punct'] = data. \
                                    apply(lambda x: \
                                    re.sub(r"([^\w\s])", \
                                    "", \
                                    x['review_no_stopwords_no_punct'] \
                                    ), axis=1)
```

```python
data.head()
```

## tokenizing

```python
data['tokenized'] = data.apply(lambda x: \
                               word_tokenize( \
                               x['review_no_stopwords_no_punct'] \
                               ), axis=1)

```

```python
data.head()
```

```python
data['tokenized'][0]
```

### Stemming

```python
ps = PorterStemmer()
```

```python
data["stemmed"] = data["tokenized"] \
                  .apply(lambda tokens: \
                  [ps.stem(token) \
                   for token in tokens])
```

```python
data.head()
```

```python
data['stemmed'][0]
```

## lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
data["lemmatized"] = data["tokenized"] \
                    .apply(lambda tokens: \
                    [lemmatizer.lemmatize(token) \
                     for token in tokens])

```

```python
data['lemmatized'][0]
```

```python
data.head()
```

```python
data.head()
```

N grams

```python
tokens_clean = sum(data['lemmatized'] , [])
```

```python
unigrams = (pd.Series \
            (nltk.ngrams(tokens_clean , 1)) \
            .value_counts())
print(unigrams)
```

```python
bigrams = (pd.Series \
           (nltk.ngrams(tokens_clean, 2)) \
           .value_counts()) 
print(bigrams)
```

```python
ngrams_4 = (pd.Series \
            (nltk.ngrams(tokens_clean, 4)) \
            .value_counts()) 
print(ngrams_4)
```
